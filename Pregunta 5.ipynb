```python
import pandas as pd
df = pd.read_csv('pregunta5.csv')
df.keys()
```

```python
print('Dataframe')
print(df)
```

```python
df = pd.get_dummies(df, columns = ["trabaja"], drop_first = True)
```

```python
viviendas = {"propia":0,"familiar":1,"arrienda":2}
df['vivienda'] = df['tipo_vivienda'].replace(viviendas)
seguros = {"si con condiciones": 1, "no":0, "si":2}
df['ofrece_seguro'] = df['ofrece_seguro'].replace(seguros)
df
```

```python
df = pd.get_dummies(df, columns = ["tiene_hijos"], drop_first = True)
df = pd.get_dummies(df, columns = ["casado"], drop_first = True)
```

```python
df.drop(['tipo_vivienda'], axis=1, inplace=True)
df.drop(['id'], axis=1, inplace=True)
```

```python
df
```

```python
import numpy as np
X = np.array(df.drop(['ofrece_seguro'],1))
y = np.array(df['ofrece_seguro'])
X.shape
```

```python
from sklearn import linear_model
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

model = linear_model.LogisticRegression(multi_class='auto')
model.fit(X,y)
```

```python
predictions = model.predict(X)
predictions
```

```python
model.score(X,y)
```

```python
# VALIDACIÓN DEL MODELO
"""
Una buena práctica en Machine Learning es la de subdividir nuestro conjunto de datos de entrada 
en un set de entrenamiento y otro para validar el modelo (que no se utiliza durante el entrenamiento 
y por lo tanto la máquina desconoce).
"""
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, y, test_size=validation_size, random_state=seed)
```

```python
name='Logistic Regression'
kfold = model_selection.KFold(n_splits=10, random_state=seed)
cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(msg)
```

```python
predictions = model.predict(X_validation)
print(accuracy_score(Y_validation, predictions))
```

```python
# MAtriz de confusión
print(confusion_matrix(Y_validation, predictions))
```

```python
print(classification_report(Y_validation, predictions))
```
